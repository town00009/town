```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Food-Truck(For Linear Regression Program).csv")
df
```

```python
x_mean=df['6.1101'].mean()
y_mean=df['17.592'].mean()
print(x_mean,y_mean)
```

```python
df['x-x_mean']=df['6.1101']-x_mean
df['y-y_mean']=df['17.592']-y_mean
df['x-x_mean^2']=df['x-x_mean']**2
df['y-y_mean^2']=df['y-y_mean']**2
df
```

```python
df['(x-x_mean).(y-y_mean)']=df['x-x_mean']*df['y-y_mean']
df
```

```python
sum_of_XmeanYmean=df['(x-x_mean).(y-y_mean)'].sum()
sum_of_XmeanYmean
```

```python
sum_of_XmeanSqr=df['x-x_mean^2'].sum()
sum_of_YmeanSqr=df['y-y_mean^2'].sum()
sum_of_XmeanSqr,sum_of_YmeanSqr
```

```python
r=sum_of_XmeanYmean/(sum_of_XmeanSqr*sum_of_YmeanSqr)**0.5
r
```

```python
n=df.shape[0]
sx=(sum_of_XmeanSqr/n-1)**0.5
sy=(sum_of_YmeanSqr/n-1)**0.5
sx,sy
```

```python
m=(r*sy)/sx
m
```

```python
c=y_mean-m*x_mean
c
```

```python
df['y_pred']=m*df['6.1101']+c
df
```

```python
df['error']=df['17.592']-df['y_pred']
df['error^2']=df['error']**2
df
```

```python
plot1=plt.scatter(df["6.1101"],df["17.592"])
plot2=plt.plot(df["6.1101"],df["y_pred"])
plt.show()
```

```python
sse=df['error^2'].sum()
sse
```

```python
sst=df['y-y_mean^2'].sum()
sst
```

```python
df['y_pred-y_mean']=df['y_pred']-y_mean
df['y_pred-y_mean^2']=df['y_pred-y_mean']**2
df
```

```python
ssr=df['y_pred-y_mean^2'].sum()
ssr
```

```python
R2=ssr/sst
R2
```
